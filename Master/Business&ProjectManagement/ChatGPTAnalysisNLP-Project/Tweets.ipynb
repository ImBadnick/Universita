```python
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

#SCRAPE TWEETS

```python
# SCRAPE TWEETS (OR RECOLLECT FROM CSV)
tweets = []
start_date = datetime.strptime("2022-12-1", '%Y-%m-%d').date()
stop_date = datetime.strptime("2023-02-07", '%Y-%m-%d').date()

while str(start_date) != str(stop_date):
    count = 0
    # setup of the query to scrape
    query_string = "#chatgpt since:" + str(start_date - timedelta(days=1)) + " until:" + str(start_date)
    print(query_string)
    for tweet in sntwitter.TwitterSearchScraper(query_string).get_items():
        if count == 2000: break
        tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
        count+=1
    
    print(str(start_date) + " complete!")
    start_date = start_date + timedelta(days=1)
    

tweets_dataframe = pd.DataFrame(tweets, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
tweets_dataframe.to_csv("snscrape_dataset.csv", sep=",", index=False)

print("done!")
```
